```python
import numpy as np

x = np.array([0, 5])
y = np.array([3, 3])
```

```python
# algebra
dproduct = np.dot(x, y)
print(dproduct)
```

```python
# geometry
x_length = np.linalg.norm(x)
y_length = np.linalg.norm(y)
cos_xy = np.cos(np.pi/4) # 45 degree

dproduct = x_length*y_length*cos_xy
print(dproduct)
```

```python
# cosine similarity

import numpy as np
from numpy import dot
from numpy.linalg import norm

x = np.array([4, 2, 1, 2])
y = np.array([1, 2, 2, 0])

cos_sim = dot(x, y) / (norm(x)*norm(y))
print(cos_sim)
```

```python
u = 2*x
v = 3*y

cos_sim = dot(u, v) / (norm(u)*norm(v))
print(cos_sim)
```

```python
from numpy.linalg import norm

def cos_sim(x, y):
    x = x.flatten()
    y = y.flatten()
    
    x = x.astype(np.float64)
    y = y.astype(np.float64)
    
    cos_sim = np.dot(x, y) / (norm(x)*norm(y))
    return cos_sim
```
