```python
import numpy as np
import time

# Vectorized version of f(x)
def f(x):
    return x**3 + np.sin(x) * np.exp(-3*x)

# Scalar (loop-based) version of f(x)
def f_scalar(x_array):
    results = []
    for x in x_array:
        results.append(x**3 + np.sin(x) * np.exp(-3*x))
    return np.array(results)

# Array sizes to test
sizes = [10, 100, 1000, 10_000, 100_000]

# Measure and compare performance
for size in sizes:
    x = np.linspace(0, 1, size)

    # Vectorized timing
    start = time.perf_counter()
    y_vec = f(x)
    end = time.perf_counter()
    t_vec = (end - start) * 1000  # convert to ms

    # Scalar timing
    start = time.perf_counter()
    y_sca = f_scalar(x)
    end = time.perf_counter()
    t_sca = (end - start) * 1000  # convert to ms

    speedup = t_sca / t_vec if t_vec > 0 else float('inf')

    print(f"{size:<10}{t_vec:<20.4f}{t_sca:<20.4f}{speedup:.1f}x")
```
