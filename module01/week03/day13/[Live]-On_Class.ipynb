# Inheritance

```python
class People:
    # Constructor
    def __init__(self, name, age, gender):
        # Attributes
        self.name = name
        self.age = age
        self.gender = gender

    # Methods
    def show_name(self):
        print(self.name)

    def show_age(self):
        print(self.age)
```

```python
class Student(People):
    def __init__(self, name, age, gender, gpa, std_id):
        super().__init__(name, age, gender) ## Common Attributes
        self.gpa = gpa
        self.std_id = std_id

    def show_gpa(self):
        print(self.gpa)

    def show_std_id(self):
        print(self.std_id)
```

```python
class Teacher(People):
    def __init__(self, name, age, gender, salary, subject):
        super().__init__(name, age, gender)
        self.salary = salary
        self.subject = subject
```

```python
std1 = Student("John", 32, "Female", 3.5, "std001")
```

```python
vars(std1)
```

```python
std1.show_name()
```

```python
class Math2:
    def is_even(self, a):
        if a % 2 == 0:
            return True
        else:
            return False

    def factorial(self, x):
        result = 1
        for i in range(1, x+1):
            result *= i
        return result

    def estimate_euler(self, n):
        result = 1

        for i in range(1, n+1):
            result = result + 1/self.factorial(i)

        return result
```

## Inheritance Practice

```python
class Math:
    def is_even(self, a):
        if a % 2 == 0:
            return True
        else:
            return False

    def factorial(self, x):
        result = 1
        for i in range(1, x+1):
            result *= i
        return result
```

```python
class MathChild(Math):
    def estimate_euler(self, n):
        result = 1

        for i in range(1, n+1):
            result = result + 1/self.factorial(i)

        return result
```

```python
math_child = MathChild()
```

```python
math_child.is_even(2)
```

```python
math_child.estimate_euler(20)
```
